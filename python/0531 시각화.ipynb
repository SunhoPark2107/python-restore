```python
# 파이그래프는 자동으로 백분율로 나눠서 그래프 그려주는데, 굳이 백분율로 나눠서 파이그래프 만드는 코드를 짜고 싶다면 다음과 같다.
data = [75,25,50,20]
data2 = []
datasum = 0
for i in data:
    datasum += i
for i in data:
    data2.append(i/datasum*100)
plt.pie(data)
plt.show()
```

```python
# 히스토그램
# randn 은 0에서부터 1까지
# rand는 더 큰 수도, 더 작은 수도 가능.
data = np.random.randn(1000)
plt.hist(data, bins = 20)           # bins => "통"이라는 뜻.(데이터를 담는 통을 의미한다고 보면 된다. 막대 범위 bins값으로 조정한다.)
                                     # bins가 너무 높으면 안 좋음.(최빈값, max값이 다르게 됨.)
                                     # 모델링에 문제가 생김.
plt.show()
```

```python
dataA = np.random.standard_normal((100, 2))
dataB = np.random.standard_normal((100, 2))
dataC = np.random.standard_normal((100, 2))
dataA += np.array((-3, -3))
dataB += np.array((2, 4))
dataC += np.array((-1, 3))
plt.scatter(dataA[:, 0], dataA[:,1], color = "#D1B2FF", edgecolor = "0")
plt.scatter(dataB[:, 0], dataB[:,1], color = "0.7", edgecolor = "0")          # 0~1 => 0은 검정, 1은 흰색. 0.5는 검정과 흰색 반반.
plt.scatter(dataC[:, 0], dataC[:,1], color = "#B2CCFF", edgecolor = "0")      # 네이버 색상표 검색해서 색상 코드 활용 가능.
plt.show()
```

```python
import matplotlib.cm as cm     # color map. 시각화를 다양한 색 요소를 사용하여 할 수 있음.
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X, Y, c = angle, cmap = cm.Spectral) # cm.Greys, cm.hsv, cm.Spectral
plt.show()
```

```python
X = np.linspace(-6, 6, 1024)          # -6부터 6까지 1024개로 쪼개기.
Y1 = np.sinc(X)
Y2 = np.sinc(X)+1
plt.plot(X, Y1, marker = 'o', color = '0.7')                   # 모든 데이터 하나하나(1024)에 다 점 찍은 것. 총 1024개으이 점 찍혔을 것.
plt.plot(X, Y2, marker = 'o', color = '0.7', markevery = 32)   # 32개의 데이터 마다 점을 찍는 것.
plt.show()
```

```python
# 점 사이즈 주는 법.
A = np.random.standard_normal((100, 2))
B = np.random.standard_normal((100, 2))
A += np.array((-1, -2))
B += np.array((0.5, 1))
plt.scatter(A[:, 0], A[: ,  1], marker = "o", color = "orange", s = 10)
plt.scatter(B[:, 0], B[: ,  1], marker = "o", s = 100)
plt.show()
```

```python
# 버블 차트
M = np.random.standard_normal((500,2))
R = np.sum(M**2, axis = 1)
plt.figure(figsize = (5, 5))            # 표 그려지는 그림 사이즈를 준 것.(5, 5면 1:1로 된다. 그래서 버블 차트 안 squre들도 다 정사각형됨.)
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.scatter(M[:, 0], M[:, 1], c = '0.7', marker = 's', s = 32*R, edgecolor = 'k')
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar = True)
radius = [1, 0.5, 0.7, 0.3, 0.1, 0.2, 0.6, 0.8, 0.5]
theta = np.linspace(0, 2*np.pi, len(radius), endpoint = False)
points = np.vstack((theta, radius)).T
plt.gca().add_patch(patches.Polygon(points, color = "0.5"))
plt.show()
```

```python
plt.axes(polar = True)
radius = [1, 0.5, 0.7, 0.3, 0.1, 0.2, 0.6, 0.8, 0.5]
radius2 = [0.5, 0.3, 0.2, 0.1, 0.7, 1, 0.1, 0.2, 0.8]

theta = np.linspace(0, 2*np.pi, len(radius), endpoint = False)
theta2 = np.linspace(0, 2*np.pi, len(radius2), endpoint = False)

points = np.vstack((theta, radius)).T
points2 = np.vstack((theta2, radius2)).T

plt.gca().add_patch(patches.Polygon(points, color = "0.5"))
plt.gca().add_patch(patches.Polygon(points2, color = "b"))

plt.show()
```
