```python
import seaborn as sns
```

```python
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)
```

```python
# 국가별 년_월에 따른 방문자_수의 변화 그래프

plt.figure(figsize=(6,6))
sns.lineplot(x='년_월', y='방문자_수', data=df_ch1, marker='o', label = '중국')
sns.lineplot(x='년_월', y='방문자_수', data=df_jp1, marker='o', label = '일본')
sns.lineplot(x='년_월', y='방문자_수', data=df_Tai1, marker='o', label = '대만')
sns.lineplot(x='년_월', y='방문자_수', data=df_USA1, marker='o', label = '미국')
sns.lineplot(x='년_월', y='방문자_수', data=df_HK1, marker='o', label = '홍콩')
sns.lineplot(x='년_월', y='방문자_수', data=df_Thai1, marker='o', label = '태국')
sns.lineplot(x='년_월', y='방문자_수', data=df_philli1, marker='o', label = '필리핀')
sns.lineplot(x='년_월', y='방문자_수', data=df_Viet1, marker='o', label = '베트남')
sns.lineplot(x='년_월', y='방문자_수', data=df_Mal1, marker='o', label = '말레이시아')
sns.lineplot(x='년_월', y='방문자_수', data=df_Sp1, marker='o', label = '싱가포르')

# 제목 추가하기
plt.title("Visitor_nation", family = "Courier New", color = 'black', weight = 'bold', size = 20)

# x축 라벨추가하기
plt.xlabel('년_월', color = 'black', weight = 'bold', size = 15)
plt.xticks(rotation=45)
# y축 라벨추가하기
plt.ylabel('방문자_수', color = 'black', weight = 'bold', size = 15)

plt.legend()
plt.show()
```
